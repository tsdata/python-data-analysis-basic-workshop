# 삼성전자 주식시세 가져오기

```python
# 삼성전자 전체 (1999년 ~ 현재)
df = fdr.DataReader('005930')
df
```

```python
# CSV로 저장
df.to_csv("samsung.csv")
```

```python
# Excel로 저장
df.to_excel("samsung.xlsx")
```

```python
# 종가 선택하기
df['Close']
```

# 주가 데이터 시각화

```python
df['Close'].plot();
```

```python
df['Close'].plot(figsize=(12, 4));
```

```python
df[['Close', 'Volume']].plot(figsize=(12, 4));
```

```python
df[['Close', 'Volume']].plot(figsize=(12, 4),
                             secondary_y=['Volume']);
```

```python
df.loc['2020-01-01':, ['Close', 'Volume']].plot(figsize=(12, 4),
                                                secondary_y=['Volume']);
```

[실습 과제] SK하이닉스의 주식시세 데이터를 수집하고, 거래량을 왼쪽 y축에 고가와 저가를 오른쪽 y축에 표시하는 시계열 차트를 생성합니다. (이때, 2023년 3월부터 6월까지 기간으로 한정)
